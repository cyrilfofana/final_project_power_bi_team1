Because of the large number of entries, a T4 GPU was used to avoid 'runtime_exceed' errors. A total of 10 iterations, with each iteration consisting of 10000 entries, were performed. The results were loaded back to the csv file

```python
# Reading the orders)revi_dataset.csv into a Dataframe
df= pd.read_csv('/content/olist_order_reviews_dataset.csv')
```

```python
df
```

## Below code is for reading the Dataframe,we read 10000 entries/iteration, therby performing 10 iterations to complete the translation of all the reviews

```python
# Code to read the'review_comment_message_column' ,translate it and load it back to the Dataframe
for i in range(90000,len(df)):
  trans = df.loc[i,'review_comment_message']
  if pd.isna(trans)== False:
    translator = Translator()
    result = translator.translate(trans)
    df.loc[i,'translated_review']= result.text
```

```python
# Writing the Dataframe to a 'csv' file
df.to_csv('aa.csv')
```
